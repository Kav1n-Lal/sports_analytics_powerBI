```python
dfs_list_bat=[df1]
dfs_list_bowl=[df2]


no_data=[]
for w in range(1,len(df)):    
    data=[]
    columns=[]
    inpt = df['TopStats'].iloc[w]
    res = ast.literal_eval((inpt))
    # print(res)
    # print(df['PlayerName'].iloc[436])
    x=sorted(list(res.keys()))
    #print(x)
    t=0
    for i in x:
        #i=='Batting & Fielding'

        a=list(res[i].keys())
        for d in a:
                columns.append(d)
        columns.append(f'player{t}')
        columns.append(i)
        t+=1

        for j in range(len(a)):
            b=list(res[i][a[j]].keys())



            z=[]
            for k in range(len(b)):
                c=res[i][a[j]][b[k]]
                z.append(c)
            
            data.append(z)
        
        if i=='Batting & Fielding' and '4s' not in a:
            data.insert(11,['0' for r in range(len(b))])
        if i=='Batting & Fielding' and '6s' not in a:
            data.insert(12,['0' for r in range(len(b))])
        if i=='Batting & Fielding' and 'BF' not in a:
            data.insert(11,['0' for r in range(len(b))])
        if i=='Batting & Fielding' and 'SR' not in a:
            data.insert(12,['0' for r in range(len(b))])
            
        if i=='Bowling' and 'Inns' not in a:
            data.insert(19,['0' for r in range(len(b))])
        if i=='Bowling' and 'BBM' not in a:
            data.insert(24,['0' for r in range(len(b))])
        if i=='Bowling' and '4w' not in a:
            data.insert(28,['0' for r in range(len(b))])
        
        data.append([df['PlayerName'].iloc[w] for f in range(len(b))])
        data.append([i for f in range(len(b))])
    print(data)                

    

    

    try:
        dfs_list_bat.append(pd.DataFrame({'Format':data[0],
         'Mat':data[1],
         'Inns':data[2],
         'NO':data[3],
         'Runs':data[4],
         'HS':data[5],
         'Ave':data[6],
         'BF':data[7],
         'SR':data[8],
         '100s':data[9],
         '50s':data[10],
         '4s':data[11],
         '6s':data[12],
         'Ct':data[13],
         'St':data[14],
         'player0':data[15],
         'Batting & Fielding':data[16]})) 

        dfs_list_bowl.append(pd.DataFrame({'Format':data[17],
         'Mat':data[18],
         'Inns':data[19],
         'Balls':data[20],
         'Runs':data[21],
         'Wkts':data[22],
         'BBI':data[23],
         'BBM':data[24],
         'Ave':data[25],
         'Econ':data[26],
         'SR':data[27],
         '4w':data[28],
         '5w':data[29],
         '10w':data[30],
         'player1':data[31],
         'Bowling':data[32]}))
    
    except:
        no_data.append(df['PlayerName'].iloc[w])
```

```python
df_bat=pd.concat(dfs_list_bat,ignore_index=True)
df_bat['nation']=[nation for i in range(len(df_bat))]
df_bat
```

```python
df_bat.to_csv(f'{cleaned}_batting_stats.csv',encoding='utf-8')
print(f'saved as {cleaned}_batting_stats.csv')
```

```python
df_bowl=pd.concat(dfs_list_bowl,ignore_index=True)
df_bowl['nation']=[nation for i in range(len(df_bowl))]
df_bowl
```

```python
df_bowl.to_csv(f'{cleaned}_bowling_stats.csv',encoding='utf-8')
print(f'saved as {cleaned}_bowling_stats.csv')
```
