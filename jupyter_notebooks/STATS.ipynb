```python
def keys_(excelsheet_name,sh_name,nation,key):  
    
    df = pd.read_excel(f'{excelsheet_name}.xlsx', sheet_name=sh_name)

    ca=[]
    for j in range(len(df)):
        inpt = df['Stats'].iloc[j]
        res = ast.literal_eval((inpt))

        try:
            s=[]
            for i in list(res[key].values()):
                s.append(dict(i)[0])
            ca.append(len(s))
        except KeyError:
            ca.append(0)
    
    max_col=list(set(ca))       
    
    
    ca_col=[]
    ca=[]
    ind=[]
    for j in range(len(df)):
        inpt = df['Stats'].iloc[j]
        res = ast.literal_eval((inpt))

        try:
            s=[]
            for i in list(res[key].values()):
                s.append(dict(i)[0])
            ca.append(s)


            s1=[]
            for i1 in list(res[key].keys()):
                s1.append(i1)
            ca_col.append(s1)
        except KeyError:
            ind.append(j)
            pass
    
    t=[i for i in range(len(df))]
    ind1=[]
    for h in t:
        if h not in ind:
            ind1.append(h)
    w=''       
    for k in ca_col:
        if len(k)==max(max_col):
            w=k
            print(f'total number of columns in {key} is:',len(w))
            print(f'which are:',w)
            break
    
    player=[]
    for v in ind1:
        player.append(list(df['PlayerName'])[v])
    len(player)
    
#     w=['Unnamed: 0', 'Span', 'Mat', 'Runs', 'HS', 'Bat Avg', '100s', 'Wkts', 'BBI', 'Bowl Avg', '5w', 'Ct', 'St', 
#    'Avg Diff','0']


    data=[]

    a=0
    while a<len(ca_col):
        z=[]

        for n in w:
            if n in ca_col[a] and n=='Unnamed: 0':
                c=ca_col[a].index('Unnamed: 0')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Span':
                c=ca_col[a].index('Span')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Mat':
                c=ca_col[a].index('Mat')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Runs':
                c=ca_col[a].index('Runs')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='HS':
                c=ca_col[a].index('HS')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Bat Avg':
                c=ca_col[a].index('Bat Avg')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='100s':
                c=ca_col[a].index('100s')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Wkts':
                c=ca_col[a].index('Wkts')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='BBI':
                c=ca_col[a].index('BBI')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Bowl Avg':
                c=ca_col[a].index('Bowl Avg')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='5w':
                c=ca_col[a].index('5w')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Ct':
                c=ca_col[a].index('Ct')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='St':
                c=ca_col[a].index('St')
                z.append(ca[a][c])

            elif n in ca_col[a] and n=='Avg Diff':
                c=ca_col[a].index('Avg Diff')
                z.append(ca[a][c])
            else:
                 z.append(np.nan)  
        data.append(z)   
        a+=1
        
    df1=pd.DataFrame(data=data,columns=w)
    df1['player']=player
    df1['nation']=[nation for i in range(len(player))]
    try:
        df1.to_csv(f'{nation}players_{key}.csv',encoding='utf-8')
    except OSError:
        key=key.replace('/','')
        df1.to_csv(f'{nation}players_{key}.csv',encoding='utf-8')
```
